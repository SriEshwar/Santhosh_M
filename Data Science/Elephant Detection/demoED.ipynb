```python
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
import os

# Load the best saved model
best_model = load_model('best_model.keras')

# Function to make predictions
def predict_image(image_path, model):
    image = load_img(image_path, target_size=(224, 224))  # Load and resize image
    img_array = img_to_array(image) / 255.0  # Normalize the image
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions for model input
    
    prediction = model.predict(img_array)[0][0]  # Get prediction
    label = 'Elephant' if prediction > 0.5 else 'Not Elephant'
    
    # Visualization
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title(f'Prediction: {label} (Confidence: {prediction:.2f})', fontsize=12)
    plt.axis('off')
    plt.colorbar()
    plt.show()
    
    return label, prediction

# Manually specify the image path
image_path = "C:/Users/ajay2/Downloads/lion.webp"

if os.path.exists(image_path):
    predict_image(image_path, best_model)
else:
    print("Invalid file path. Please check and try again.")
```
